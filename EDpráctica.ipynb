```python
import pandas as pd

# Leer archivo Excel directamente
tabla = pd.read_excel("Base de datos.xlsx", sheet_name=0, index_col=0)

# Mostrar primeras filas
print(tabla.head(4))
```

```python
# Máximo y mínimo de la columna "edad"
edad_min = tabla["edad"].min()
edad_max = tabla["edad"].max()

print("Edad mínima:", edad_min)
print("Edad máxima:", edad_max)
```

```python
import matplotlib.pyplot as plt

plt.hist(tabla["edad"], bins=20, color="darkcyan", edgecolor="black")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de edades")
plt.show()
```

```python
# Valores únicos de la columna
grupos_unicos = tabla["grupo_de_edad"].unique()
print("Grupos únicos:", grupos_unicos)
print("Cantidad de grupos:", len(grupos_unicos))

# Conteo de cada grupo
conteo_grupos = tabla["grupo_de_edad"].value_counts()
print("\nFrecuencia por grupo:")
print(conteo_grupos)
```

```python
# Contar y ordenar
conteo = tabla["grupo_de_edad"].value_counts().sort_values(ascending=False)

# Gráfico de barras
conteo.plot(kind="bar", color="darkcyan", edgecolor="black")
plt.xlabel("Grupo de edad")
plt.ylabel("Frecuencia")
plt.title("Distribución por grupo de edad")
plt.xticks(rotation=90)  # Para que no se encimen las etiquetas
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegurarte de que la columna sea tipo fecha
tabla["fecha_muerte"] = pd.to_datetime(tabla["fecha_muerte"], errors="coerce")

# Crear una nueva columna con el año-mes
tabla["mes_muerte"] = tabla["fecha_muerte"].dt.to_period("M")

# Contar muertes por mes
conteo_meses = tabla["mes_muerte"].value_counts().sort_index()

# Gráfico de barras (histograma por meses)
conteo_meses.plot(kind="bar", color="darkcyan", edgecolor="black", figsize=(12,6))
plt.xlabel("Mes")
plt.ylabel("Número de muertes")
plt.title("Distribución de muertes por mes")
plt.xticks(rotation=45, ha="right")
plt.show()
```
