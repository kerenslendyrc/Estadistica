```python
import pandas as pd
import numpy as np
df = pd.read_excel("Base de datos.xlsx")

df_work = df.copy()
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from scipy.stats import chi2_contingency
import math

target_vars = [
    "edad", "grupo_de_edad", "asignacion_enfermeria",
    "Servicio", "uo_medico", "sexo", "EPS", "departamento", "ciudad",
    "id_discapacidad", "nombre_diagnostico_egreso", "gran_causa_morbilidad",
    "causa_basica_muerte", "fecha_muerte"
]

df_work = df.copy()


presentes = [c for c in target_vars if c in df_work.columns]
if len(presentes) == 0:
    raise ValueError("Ninguna de las variables objetivo está en el DataFrame.")

#Transformaciones necesarias para χ²

# 1) edad -> categórica (manteniendo el nombre 'edad')
if "edad" in presentes and np.issubdtype(df_work["edad"].dtype, np.number):
    df_work["edad"] = pd.cut(
        df_work["edad"],
        bins=[-1, 9, 19, 29, 39, 49, 59, 69, 79, 120],
        labels=["0-9","10-19","20-29","30-39","40-49","50-59","60-69","70-79","80+"]
    ).astype("string")

# 2) fecha_muerte -> mes AAAA-MM (manteniendo el nombre 'fecha_muerte')
if "fecha_muerte" in presentes:
    df_work["fecha_muerte"] = pd.to_datetime(df_work["fecha_muerte"], errors="coerce")
    df_work["fecha_muerte"] = df_work["fecha_muerte"].dt.to_period("M").astype(str)

# 3) Asegurar string/categórico en todas las demás
for c in presentes:
    if c != "edad" and c != "fecha_muerte":
        df_work[c] = df_work[c].astype("string")

# ========= 2) FUNCIONES AUXILIARES =========
def cramers_v(chi2, n, k_rows, k_cols):
    k = min(k_rows - 1, k_cols - 1)
    if k <= 0 or n <= 0:
        return np.nan
    return math.sqrt(chi2 / (n * k))

def contingency_c(chi2, n):
    if n <= 0:
        return np.nan
    return math.sqrt(chi2 / (chi2 + n))

def phi_coef(chi2, n, k_rows, k_cols):
    if k_rows == 2 and k_cols == 2 and n > 0:
        return math.sqrt(chi2 / n)
    return np.nan

# ========= 3) CHI² Y COEFICIENTES PAR-A-PAR (solo variables solicitadas) =========
results = []
for a, b in combinations(presentes, 2):
    sub = df_work[[a, b]].dropna()
    if sub.empty:
        continue

    ct = pd.crosstab(sub[a], sub[b])

    # Requiere al menos 2 filas y 2 columnas
    if (ct.sum().sum() == 0) or (ct.shape[0] < 2 or ct.shape[1] < 2):
        continue

    # Corrección de Yates solo para 2x2 (opcional)
    use_yates = (ct.shape[0] == 2 and ct.shape[1] == 2)

    try:
        chi2, p, dof, expected = chi2_contingency(ct, correction=use_yates)
    except ValueError:
        continue

    n = ct.values.sum()
    v = cramers_v(chi2, n, ct.shape[0], ct.shape[1])
    c = contingency_c(chi2, n)
    phi = phi_coef(chi2, n, ct.shape[0], ct.shape[1])

    results.append({
        "var1": a,
        "var2": b,
        "n": int(n),
        "tabla_dim": f"{ct.shape[0]}x{ct.shape[1]}",
        "chi2": chi2,
        "gl": dof,
        "p_valor": p,
        "C_contingencia": c,
        "V_cramer": v,
        "Phi_2x2": phi
    })

tabla_coeficientes = pd.DataFrame(results).sort_values(["var1","var2"]).reset_index(drop=True)
```

```python
tabla_coeficientes
```

```python
# Filtrar solo las relaciones significativas (p < 0.05)
signif = tabla_coeficientes[tabla_coeficientes["p_valor"] < 0.05]

# Ordenar por valor de chi² (mayor primero)
ranking = signif.sort_values("chi2", ascending=False)

# Mostrar top 20 relaciones más fuertes
print("Top 20 pares de variables más relacionadas (según χ² y p<0.05):")
print(ranking[["var1", "var2", "chi2", "gl", "p_valor"]].head(20))
```

```python
vars_list = sorted(set(tabla_coeficientes["var1"]).union(set(tabla_coeficientes["var2"])))
mat = pd.DataFrame(np.nan, index=vars_list, columns=vars_list, dtype=float)

# Rellenar valores
for _, row in tabla_coeficientes.iterrows():
    a = row["var1"]
    b = row["var2"]
    v = row["V_cramer"]
    mat.loc[a, b] = v
    mat.loc[b, a] = v

# Opcional: poner 1.0 en diagonal o 0 (prefiero 1.0 para referencia visual)
np.fill_diagonal(mat.values, 1.0)

# Mostrar la matriz (pivot)
print("\nMatriz (pivot) de V de Cramér:")
display(mat)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Preparar máscara superior para que se vea solo triángulo inferior (evita duplicados)
mask = np.zeros_like(mat, dtype=bool)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(12, 10))
sns.set(style="white")
# Usar cmap por defecto de seaborn/matplotlib; anotar con 2 decimales si no es muy grande
sns.heatmap(mat.astype(float),
            mask=mask,
            annot=True,
            fmt=".2f",
            linewidths=.5,
            square=True,
            cbar_kws={"shrink": .7},
            vmax=mat.max().max() if not np.isnan(mat.max().max()) else None)

plt.title("Matriz de V de Cramér (solo triángulo inferior mostrado)")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

mat_plot = mat.fillna(0).copy()

dist = 1.0 - mat_plot.values

from scipy.spatial.distance import squareform
condensed = squareform(dist, checks=False)
Z = linkage(condensed, method='average')   # método 'average' es razonable

plt.figure(figsize=(12, 5))
dn = dendrogram(Z, labels=mat_plot.index.tolist(), leaf_rotation=45, leaf_font_size=10)
plt.title("Dendrograma (dist = 1 - V_cramer)")
plt.tight_layout()
plt.show()

order = dn['leaves']
ordered = mat_plot.iloc[order, :].iloc[:, order]

plt.figure(figsize=(10, 10))
sns.heatmap(ordered, annot=True, fmt=".2f", square=True, cbar_kws={"shrink":.7})
plt.title("Heatmap V de Cramér (ordenado por clustering)")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

v_threshold = 0.5
dist_threshold = 1.0 - v_threshold

clusters = fcluster(Z, t=dist_threshold, criterion='distance')
cluster_df = pd.DataFrame({"variable": mat_plot.index, "cluster": clusters})

representantes = []
for cl in sorted(cluster_df["cluster"].unique()):
    vars_in_cl = cluster_df[cluster_df["cluster"] == cl]["variable"].tolist()
    if len(vars_in_cl) == 1:
        representantes.append((cl, vars_in_cl[0], vars_in_cl))
        continue
    sub = mat_plot.loc[vars_in_cl, vars_in_cl]
    avg_v = sub.mean(axis=1)
    rep = avg_v.idxmax()
    representantes.append((cl, rep, vars_in_cl))

cluster_df_sorted = cluster_df.sort_values("cluster")
print("\nClusters encontrados (umbral V >= {:.2f}):".format(v_threshold))
print(cluster_df_sorted.to_string(index=False))

print("\nRepresentante propuesto por clúster (variable con mayor V promedio dentro del clúster):")
for cl, rep, members in representantes:
    print(f"Cluster {cl}: representante = {rep} | miembros = {members}")
```
